```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv('Datasets/tracks.csv')
df_tracks.head()
```

```python
most_popular = df_tracks.query('popularity>90', inplace=False).sort_values('popularity',ascending=False)
most_popular[:10]
```

- Finding if the wrong date format is present

```python
yyyy_mm_format_count = df_tracks['release_date'].apply(lambda x: len(str(x)) == 7 and '-' in str(x)).sum()

print(f"Number of dates in 'YYYY-MM' format: {yyyy_mm_format_count}")
```

- if present
- change it into the correct format like "YYYY-MM-DD"

```python
def convert_yyyy_mm_to_yyyy_mm_dd(date_str):
    # If the date is in "YYYY-MM" format (length of 7), append "-01"
    if len(str(date_str)) == 7 and '-' in str(date_str):
        return date_str + "-01"
    return date_str

# Apply the conversion to the release_date column
df_tracks['release_date'] = df_tracks['release_date'].apply(convert_yyyy_mm_to_yyyy_mm_dd)

# Convert the entire release_date column to datetime format to standardize
df_tracks['release_date'] = pd.to_datetime(df_tracks['release_date'], errors='coerce')

# Display the updated DataFrame to verify
print(df_tracks.head())
```

- Changing index to the relese date and converting into datetime

```python
df_tracks.set_index("release_date",inplace=True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

- changing the duration in milliseconds to seconds 

```python
df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x: round(x / 1000))
df_tracks.drop("duration_ms", inplace=True, axis=1)
```

```python
df_tracks.duration.head()
```

# Using correlation 

- The Pearson correlation coefficient measures the linear relationship between two variables, ranging from -1 to +1, where: +1 indicates a perfect positive linear relationship, -1 indicates a perfect negative linear relationship, 0 indicates no linear relationship.

```python
corr_df = df_tracks.drop(columns=["key", "mode", "explicit"], errors='ignore')
corr_df = corr_df.select_dtypes(include=[np.number]).corr(method="pearson")
```

- using matplotlib and seaborn

```python
from matplotlib.pyplot import figure
figure(figsize=(14,7),dpi=90)
```

- Heatmaps are a popular data visualization technique that uses color to represent different levels of data magnitude,

- annot : If set to True , the value in each cell is annotated on the heatmap.

```python
heatmap = sns.heatmap(
    corr_df,annot=True,fmt=".1g",
    vmin=-1,vmax=1,center=0,
    cmap="inferno",linewidths=1,linecolor="Black")
heatmap.set_title("Correlation HeatMap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

- creating a regression plot b/w loudness and energy
- taking a sample dataset from 0.004% of the original dataset

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
figure(figsize=(14,7))
sns.regplot(
    data = sample_df,
    y = "loudness",
    x = "energy",
    color= 'c'
).set(title="Loudnes VS Energy Correlation")
```
